```python
import numpy as np
import matplotlib.pyplot as plt
import torch
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu6(x):
    return np.minimum(np.maximum(0, x), 6)

def relu8(x):
    return np.minimum(np.maximum(0, x), 8)

def relut(x):
    return np.minimum(np.maximum(0, x), 6.375)
def f(x):
    return x * sigmoid(x)

def g(x):
    return x * relu6(x + 3) / 6

def h(x):
    return x * relu8(x + 4) / 8

def y(x):
    return x * relut(x + 3.1875) * 0.15625

def func(x):
    input_clamp = torch.clamp(x + 3.1875, 0, 6.375) * 0.15625
    return x * input_clamp

# 定义x范围
x = np.linspace(-10, 10, 1000)

# 计算y值
y_f = f(x)
y_g = g(x)
y_h = h(x)
y_y = y(x)

# 绘制图像
plt.figure(figsize=(10, 6))
plt.plot(x, y_f, label='f(x) = x * sigmoid(x)')
plt.plot(x, y_g, label='g(x) = x * relu6(x + 3) / 6')
plt.plot(x, y_h, label='h(x) = x * relu8(x + 4) / 8')
plt.plot(x, y_y, label='y(x) = x * relu6(x + 3) * 0.15625')
x = torch.linspace(-10, 10, 1000)  # 在-10到10之间均匀地取1000个点
y = func(x)  # 计算对应的y值
x = x.numpy()
y = y.numpy()
plt.plot(x, y, label='func(x)')
plt.title('Plot of f(x), g(x), h(x) and y(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
